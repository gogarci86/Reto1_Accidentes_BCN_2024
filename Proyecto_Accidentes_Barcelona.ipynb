### 2. Limpieza de Datos (Data Cleaning)

#### 2.1 Gestión de duplicados

Cuento todos los duplicados y los muestro por pantalla, para chequear potenciales errores.

```python
accidentes.duplicated(keep=False).sum()
```

Hay un total de 4 registros que parecen duplicados. Tienen distinto numero de expediente pero los datos del accidente son iguales (coinciden en dia y fecha con el único cambio del nº de expediente). Voy a asumir que son debidos a duplicidades en la toma de datos, con lo que me quedaré solo con el primer valor que aparece. Hago la comprobación de que no hay duplicados tras eliminarlos y listo el DF para ver nº de registros (4 menos que el aterior 7536 vs 7532)

```python
accidentes[accidentes.duplicated(keep=False)]
```

```python
accidentes.drop_duplicates(keep='first', inplace=True)
accidentes.duplicated().sum()
```

#### 2.2 Conversión de tipos de datos

Tras analizar los tipos de datos, tiene sentido incluir desde código de distrito hasta descripción de causa como variables categóricas y dejar el resto como variables numéricas. Procedo a cambiar el tipo de variable según lo comentado, y lo compruebo por pantalla

```python
col=accidentes.loc[:,'Codi_districte':'Descripcio_causa_vianant'].columns
accidentes[col]=accidentes[col].astype('category')
accidentes.dtypes
```

```python
accidentes.describe(include='category').T
```

#### 2.3 Gestión de valores nulos o faltantes

A continuación saco un listado de valores nulos y lo ordeno de manera descendente

```python
accidentes.isna().sum().sort_values(ascending=False)
```

Para la eliminación de nulos asumiremos los siguientes supuestos:
1. **Número de muertos, Número de lesionados graves y número de lesionados leves**: los nulos los sustituiremos por el valor 0. Tiene sentido que la guardia urbana haya metido N/A en el caso de que no haya muertos ni lesionados de ninguna consideración
2. **Num postal, descripción de causa vianant**: crearemos una nueva categoría 'No disponible' y cambiaremos los nulos por valor 'No disponible'
3. **Numero_victimes**: Tiene sentido remplazar el valores nulos por el valor que mas se repite (moda). No reemplazo estos valores por 0 porque siempre en un accidente hay victimas (al menos 1).
